```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

```python
img = image.load_img('dog/dog.png', target_size=(200, 200))
```

```python
img
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'  # nearest, constant, reflect, wrap
)
```

```python
type(img)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
input_img = img.reshape(1, 200, 200, 3)
```

```python
i=0

for batch in datagen.flow(input_img, batch_size=1, save_to_dir='dog', save_prefix='dog', save_format='png'):
    i += 1
    if i == 10:
        break
```
